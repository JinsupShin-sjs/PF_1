```python
#사용자 영문 이름 입력 받아서  성과 이름 순서를 바꾸어
#출력하는 프로그램 작성
#성 이름 공백으로 구분 인풋값 : hong gildong
#hong
#gildong
full_name = input("성과 이름을 공백을 구분하여 넣어주세요:")
print(full_name)
```

```python
space = full_name.find(' ')
print(space)
```

```python
first_name = full_name[:space]
print(first_name)
```

```python
first_name = full_name[:space]
last_name = full_name[space+1:]
print(first_name)
print(last_name)
```

```python
full_name.capitalize() #<-첫글짜만 대문자로 바꾸는 함수
```
