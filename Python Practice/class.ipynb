* 속성 : brasnd, model, coloir
* 기능 : turn_on, turn_off, drive

```python
class Car:
    Car_count = 0 
    def __init__(self, name, cc, year):
        self.__name = name
        self.__cc = cc
        self.__year = year
        Car.car_count +=1

    def check(self):
        if self.__cc < 1000:
            print ('소형')
        elif 1000 <= self.__cc <= 2000:
            print ('중형')    
        elif 2000 < self.__cc: 
            print ('대형')    
```

```python
class Car:
    def __init__(self, brand, model, color):
        self.brand = brand
        self.model = model
        self.color = color
        print(self.brand, self.model, self.color ,'출고')

    def turn_on(self):
        print(self.brand, '시동울 켭니다.')
    def turn_off(self):
        print(self.brand, '시동울 끕니다.')
    def drive(self):
        print(self.brand, '주행합니다.')
```

```python
car1 = Car('현대', '소나타','화이트')
```
