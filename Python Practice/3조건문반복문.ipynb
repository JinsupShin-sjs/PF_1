# 1.1 표준입력 장치를 활용한 if 문

```python
#점수를 입력받아서, 점수가 60점 이상이면 합격을 출력 아니면 불합격
score = int(input('점수를 입력하세요'))
if score >= 60:
    print('합격')
    print('축하한다')
print('수고하셨습니다.')
```

```python
score = int(input('점수를 입력하세요'))
if score >= 60:
    print('합격')
else:
    print('불합격')
print('수고하셨습니다.')
```

```python
# if elif else 를 사용하여 점수를 입럭벋아서 90 이상 A 80 이상 B
# 70점 이상이면 C 나머지는 F로 
score = int(input('점수를 입력하세요'))
if score >=90: 
    grade = 'A'
elif score >=80: 
    grade = 'B'
elif score >=70: 
    grade = 'c'
else:
    grade = 'F'
print('grade : ', grade)
```

# 2. 반복문 (While문) 
* 조건을 만족할 때 까지 계속 수핼

```python
save_money = 0
while save_money < 12:
      save_money += 1 # save_money += #save_money = +1
      print('적금을 %d번 넣었습니다. ' %save_money)
      if save_money == 12:
          print('적금을 받았습니다')
```

```python
#
food = 100
money = 12000

while money:
    print('돈을 내고 음식을먹습니다.')
    food = food - 10
    print('남은 음식의 양은 %d gram입니다.' %food)
    if not food:
          print('음식이 없어 먹을 수 없습니다.')
          break
```

```python
# 퀴즈
# 1. 숫자를 입력 받아서 3의 배수인지 아닌지를 판단하는 프로그램을 만드시오
# 00은 3의 배수입니다.
num = int(input('숫자를 입력하시요'))
if num % 3 == 0:
    print(f'{num}은 3의 배수입니다')
else:
    print(f'{num}은 3의 배수가 아닙니다')
```

```python
# 퀴즈2. 1~100 까지 더하는 코드를 while을 활용하여 만드시오
i = 0
result = 0
while i < 100:
    i = i + 1
    result = result + i
print(result)
    #if num = 100
    #break
```

```python
# 퀴즈 3. 1~100 까지 짝수만 더하는 코드를 while을 활용하여 만드시오
i = 0
result = 0
while i < 100:
    i = i + 1
    if i % 2 == 0:
        result = result + i
print(result)
```
