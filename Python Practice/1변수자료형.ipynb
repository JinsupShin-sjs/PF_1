# 2.4 관계연산자

```python
# (1) 동등 비교
bool_result = num1 == num2
print (bool_result)
# 100 == 100,  100 != 20
bool_result2 = num1 != num2
print (bool_result2)
```

```python
#(2) 크기비교
bool_result3 = num1 > num2 #num1 값이 큰지 비교
print(bool_result3)

bool_result4 = num1 < num2 #num1 값이 작은지 비교
print(bool_result4)

bool_result5 = num1 >= num2 #num1 값이 크거나 같은지 비교
print(bool_result5)

bool_result6 = num1 <= num2 #num1 값이 작거나 같은지 비교
print(bool_result6)
```

# 2.9 format 과 양식문자

```python
'''
%양식문자
%s : 문자열
%c : 한개의 문자
%d : 정수
%f : 실수
%o : 8진수
%x : 16진수
%% : %출력
'''
```

```python
 print('원주율', format(3.14159, "8.4f")) 
 print('금액', format(10000, "8,d"))
```

```python
name = "홍길동"
age = 25
price = 125.456
print(name, age, price)
print('이름 : %s, 나이 : %d, data = %.2f' %(name, age, price))
print('이름 : {},나이 : {}, data ={}' . format(name, age, price))
```

```python
#축약형
print(f'이름:{name}, 나이:{age}. data:{price}')
print('원주율', format(3.14159, ".2f"))
print(f'원주율 : {3.14159:.2f}')
x=3.14159
print(f'원주율 : {x:.2f}')
```

```python
print('금액', format(10000, ",d")) 
print(f'금액:{10000:,}')  
```

```python
원주율 = 3.14159
print("원주율 =", format(3.14159, "8.3f")) #HW 교재 46~47 4가지로 해보기
print(f'원주율 = {3.14159: 8.3f}')
print('원주율 = %s. 2f,' %(원주율))
```
