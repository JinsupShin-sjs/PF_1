```python
#a[1] = 'b'
# 딕셔너리명 = dict(키1=값1, 키2=값2, ...)
# 키값에 ''("")를 쓰지 않는다
menu1 = dict(김밥=3000, 떡볶이=2500, 어묵=2000, 튀김=3000)
menu1
```

```python
# 딕셔너리명 = dict(zip(key리스트, value리스트))
key_list = ['김밥', '떡볶이', '어묵', '튀김']
value_list = [2000, 2500, 3000, 3000]
menu2 = dict(zip(key_list, value_list))
menu2
```

```python
#딕셔너리 키가 존재하지 않으면 추가 존재 하면 수정
scores = {'kor':100, 'eng':90, 'math':80}
scores['math'] = 85
scores
```

```python
scores['music'] = 95
scores
```

```python
# setdefault(key, [default])
# key가 딕셔너리에 있으면 해당값을 반환
# 없으면 default값을 갖는 key를 삽입한 후에 default 반환함
# default 값은 None
scores = {'kor':100, 'eng':90, 'math':80}
scores.setdefault('kor')
```

```python
scores = {'kor':100, 'eng':90, 'math':80}
scores.setdefault('music')
scores
```

```python
scores = {'kor':100, 'eng':90, 'math':80}
scores.setdefault('music', 100)
scores
```

```python
scores.setdefault('music', 90)
scores
```

```python
# update로 여러 항목을 추가/수정 할 수 있음
# 키가 존재하면 수정되고 존재하지 않으면 추가
# 딕셔너리명.update(키1=값1, 키2=값2)
scores = {'kor':100, 'eng':90, 'math':80}
scores.update(math=90, music=90)
# scores.update(zip(['math', 'music'], [90,90]))
# scores.update([('math', 90),('music', 90)])
```

```python
scores
```

```python
# 딕셔너리에서 항목 삭제
del scores['kor']
```

```python
scores
```

```python
#딕셔너리명.pop(키, 기본값)
math = scores.pop('math')
```

```python
scores
```

```python
math
```

```python
scores.clear()
scores
```

```python
이름 = input('이름: ')
나이 = input('나이: ')

개인정보 = {'이름' : 이름, '나이' : 나이}

for k in 개인정보.keys():
    print('-키:', k)

for v in 개인정보.values():
    print('-키:', v)
```

```python
# enumerate()
t = [1, 5, 7, 33, 39]
for i in enumerate(t):
    print(i)
```

```python
for i, v in enumerate(t):
    print(f'index : {i}, value: {v}')
```

```python
x = {'a':10, 'b':20, 'c':30, "d":40}
for i, (k, v)  in enumerate(x.items()):
    print(i, k, v)
```
