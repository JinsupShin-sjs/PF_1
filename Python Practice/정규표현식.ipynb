```python
import re
```

```python
# . 은 한개의 임의 문자임
r = re.compile ('a.c')
r.search('kkk')
```

```python
r.search('abc')
```

```python
# ? 는 ?앞애 문자기 존재할 수도 있고 아닐 수도 있다 
r = re.compile('ab?c')
r.search('abbc')
```

```python
r.search('abc')
```

```python
#  * 는 바로 앞의 문자디 0개 이상일 경우
r = re.compile('ab*c')
r.search('a')
```

```python
 r.search('abbbc')
```

```python
 r.search('ac')
```

```python
 # + 는 앞의 문자가 최소 1개 이상 
 r = re.compile('ab+c')
r.search('ac')
```

```python
 r.search('abbbc')
```

```python
#  ^ 시작되는 문자열
r = re.compile('^ab')
r.search('zab') 
```

```python
r.search('abz')
```

```python
#  {숫자}, 해당 문자를 숫자 만큼 반복
r = re.compile('ab{2}c')
r.search('abbc') 
```

```python
# {숫자1, 숫자2} 해당 문장를 숫자1 이상 숫자 2 이하
r = re.compile('ab{2,8}c')
r.search('abc')
```

```python
r.search('abbbbbbbbc')
```

```python
# [] 안에 믄지들을 널으면 그문자들 중에서 한대의 문자와  매치 하라는 의미
r = re.compile('[abc]')
r.search('zzz')
```

```python
 r.search('a')
```

```python
 r.search('aaaaaaa')
```

```python
r.search('baac')
```

```python
#  알파벳 소문자
r = re.compile('[a-z]')
r.search('aBC') 
```

```python
r.search('111')
```

```python
r.search('ABC')
```

```python
r = re.compile('[a-zA-Z]')
r.search('1B3')
```

```python
 # [^문자] 는 ^뒤에 붙은 문자들을 제외한  모든 문자를 매치하는  역할
 r = re.compile('[^abc]')
r.search('a')
```

```python
r.search('ab')
```

```python
r.search('d')
```

```python
text3 = '''100 Jhon PROF
101 James STUD
102 Mac STUD'''
```

```python
re.findall('[a-zA-Z1-9]+', text3) # 숫자 X
```

```python
#split()	정규 표현식을 기준으로 문자열을 분리하여 리스트로 리턴합니다.
text3.split()
```

```python
re.split('\s+', text3) #\S	공백을 제외한 문자를 의미합니다. [^ \t\n\r\f\v]와 의미가 동일합니다.
```

```python
# 1. d 사용 방식
# 2 [a-z] 방식으로 정수만 가져오기
re.findall('\d+', text3)
re.findall('[0-9]+' , text3)
```

```python
#대문자가 4개 익상인 문자를 첮우세요 PROF STUD
re.findall('[A-Z]{4}' , text3)
#첫글자가 대문자이후 소문자 인 아이들 
re.findall('[A-Z][a-z]+' , text3)
```
