```python
#min max
import random
dataset = []
for i in range(10):
    r = random.randint(1,100)
    dataset.append(r)
print(dataset)
```

```python
#변수 초기화 / -> 알고리즘의 기초!!
vmax = vmin = dataset[0]

for i in dataset:
    if vmax < i:
        vmax = i
    if vmin > i:
        vmin = i
print(f'최대값{vmax}, 최소값{vmin}')
```

```python
#입력 문잘열룰 한중애 다섯 굴자씩 출력하는 print_5xn(string) 함수를 만드세요
#'아이엠어보이유알어걸'
# 출력 '아이엠어보'
#'이유알어걸'
```

```python
def print_5xn (line):
    check_num = int(len(line) / 5)
    for x in range(check_num +1):
        print(line[x*5: x*5 + 5])
```

```python
 print_5xn('아이엠어보이유알어걸')
```
