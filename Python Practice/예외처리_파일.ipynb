```python
a = [1,2,3]
try:
    x = int(input()) 
    y = 100 / x
    print(y, a[x])
except ValueError:
    print("입력 값 잘못")
except ZeroDivisionError:
    print('0으로 못나눔')
except IndexError:
    print("Index 에러발생")
except Exception as e:
    print(e)
```

# ELSE - finally

```python
try:
    x = int(input())
    y = 100 / x
except Exception as e:
    print(e)
else: #try 코드가 예외가 발생하지 않을때 실행함
    print(y)
```

```python
try:
    x = int(input())
    y = 100 / x
except Exception as e:
    print(e)
else: 
    print(y)
finally:
    print("코드실행 종료")
```

```python
def three_multiple():
    try:
        x = int(input("3의 배수를 입력하세요: "))
        if x %3 != 0:
            raise Exception('3의 배수가 아닙니다')
    except Exception as e:
        print('three_multiple 함수에서 에러가발생했습나다,', e) 
        raise
```

```python
try:
    three_multiple()
except Exception as e:
    print('스크립크에서 에러가 발생:', e)
```

*assert로 예외발생
- assert 는 지정된 조건식이 거짓일 떼 assertError 를 발생시 조건 값이 참이면 그냥 넘어감
- assert 가 나와서눈 앉란ㄴ 조건을 검사항ㄹ 짜 강ydfja
- 예를 들면 3의 배수가 아니면 예외를 방생, 3의 배수면 그냔 넘어감

```python
x = int(input('3의 배수를 입력'))
assert x %3 == 0, '3의 배수가 아닙니다.'
print(x)
```

```python
try:
    x = int(input('3의 배수를 입력'))
    assert x %3 == 0, '3의 배수가 아닙니다.'
except Exception as e:
    print(e)
else:
    print(x)
```
