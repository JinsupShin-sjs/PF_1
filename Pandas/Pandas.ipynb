```python
# 리스트를 데이터 프레임으로
list_data = [[15 ,'남','신촌중'] , [17, '여', '강남중']]
df2 = pd.DataFrame(list_data, index=['길동', '영희']), columns=['나이','성별','소속']
print(df2)
print(df2.index)
print(df2.columns)
```

```python
#리스트의 행과 열 이름 바꾸기
# 행 : 길동, 영희를 -> 학생1, 학생2
# 열 : 나이 성별 소속 -> 연련 , 남여, 학교 
df2.index = ['학생1', '학생2']
df2.columns = ['연령', '남여', '학교']
print(df2)
```

```python
#행과 열 이름을  바꾸는 다른 방법
df2.rename(columns = {'연령':'나이', '남여':'성별, '학교':'소속'}) # 바궈지지만 적용안됨
```

```python
print(df2) #바뀌지 않음
# inplace 로 적용
df2.rename(columns = {'연령':'나이', '남여':'성별, '학교':'소속'}), inplace=True
print(df2)
```

```python
#학생1 준서, 학생2 예은

df2.rename(index) = {'학생1':'준서', '학생2':'예은'}), inplace=True 
print(df2)
```
