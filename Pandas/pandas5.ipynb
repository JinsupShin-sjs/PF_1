```python
import numpy as np
import pandas as pd
df1=pd.DataFrame({'a':['a0', 'a1', 'a2', 'a3'],'b':['b0', 'b1', 'b2', 'b3'],'c':['c0', 'c1', 'c2', 'c3']}, index=[0,1,2,3])
```

```python
result1 = pd.concat([df1, df2]) 
```

```python
#열방향으로 이어 붙이기
```

```python
result3  = pd.concat([df1, df2], axis=1)
result3
```

```python
#join = 'inner' 옵션적용하기 (교집합)
result3_in  = pd.concat([df1, df2], axis=1, join='inner')
result3_in
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex','class', 'fare', 'survived']]
```

```python
print('승객수:', len(df))
```

```python
#Class 열을 기준으로 분할
grouped = df.groupby(['class'])
print(grouped)
```

```python
#그룹 객체를 iteration 으로출력
for key, group in grouped:
    print('+key:', key)
    print('+number:', len(group))
    print(group.head())
    print('\n')
```

```python
#연산메소드 적용
average = grouped.mean()
print(average)
```

```python
#개별 그룹 선택
group3 = grouped.get_group('Third')
print(group3.head())
```

```python
grouped_two = df.groupby(['class', 'sex'])
print(grouped_two.head())
```

```python
for key, group in grouped_two:
    print('*key:', key)
    print('*number:', len(group))
    print(group.head())
    print('\n')
```

```python
average_two = grouped_two.mean()
```

```python
average_two
```

```python
group_3f=grouped_two.get_group(('Third','female'))
print(group_3f.head())
```

```python
def min_max(x):
    return x.max() - x.min()
```

```python
agg_minmax = grouped.agg(min_max)
agg_minmax.head(3)
```

```python
agg_all = grouped.agg(['min', 'max'])
agg_all.head(3)
```

```python
agg_sep = grouped.agg({'fare':['min', 'max'], 'age': 'mean'})
agg_sep.head(3) 
```

```python
#내가만든 함수 적용해서 값을 바꿔보가
def peak_to_peak_ratio(x):
    return x.max() / x.min()
```

```python
iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
iris.groupby(iris.species).agg(lambda x : x.max() / x.min())
```

```python
iris.groupby(iris.species).apply(lambda x : x.max() / x.min())
```
