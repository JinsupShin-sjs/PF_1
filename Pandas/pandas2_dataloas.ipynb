scatter -stripplot, swarmplot

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure (figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.srtipplot(x='class', y='age', data=titanic, ax=ax1)
sns.swarmplot(x='class', y='age', data=titanic, ax=ax2)
ax1.set_title('strip')
ax2.set_title('swarm')
plt.show()
```

막대그래프

```python
fig = plt.figure (figsize= (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1)
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2) 
sns.barplot(x='sex', y='survived', hue='class', dodge = False, data=titanic, ax=ax3) 

plt.show()
```

countplot

```python
fig = plt.figure (figsize= (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
sns.countplot(x='class', palette='Set1', data =titanic, ax=ax1)
sns.countplot(x='class', hue='who', palette='Set2' , data =titanic, ax=ax2)
sns.countplot(x='class', hue='who', palette='Set2' , dodge=False, data = titanic, ax=ax3)
```

boxplot

```python
fig = plt.figure (figsize= (15,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)
sns.boxplot (x='alive', y='age',data=titanic, ax=ax1) # box 기본  
sns.boxplot (x='alive', y='age',hue = 'sex',  data=titanic, ax=ax2) # box hue   
sns.violinplot (x='alive', y='age',data=titanic, ax=ax3) # violin 기본
sns.violinplot (x='alive', y='age',hue = 'sex',  data=titanic, ax=ax4) # violin hue   
plt.show()
```

```python
sns.set_style('darkgrid') # whitegrid, white, dark, ticks, darkgrid
fig = plt.figure(figsize=(15,5))
#3개의 써브플롯
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)
#displot
sns.displot(titanic['fare'], ax=ax1)
#kdeplot
sns.kdeplot(x ='fare', data=titanic, ax=ax2)
#histoplot
sns.histplot(x ='fare', data=titanic, ax=ax3)
ax1.set_title('t-displot')
ax2.set_title('t-kdeplot')
ax3.set_title('t-histplot')
plt.show()
```
